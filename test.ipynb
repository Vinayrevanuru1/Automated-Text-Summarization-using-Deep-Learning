```python
with open ("out.txt", "r") as f:
    #read data from file
   text =  f.readlines()
```

```python
text
```

```python
import pandas as pd
```

```python
df =  pd.DataFrame(columns= ['Review','Original','Predicted'])
```

```python
#add the lines to the the review column if the line starts with review
#add the lines to the the orginal column if the line starts with orginal
#add the lines to the the predicted column if the line starts with predicted
review = []
original = []
predicted = []
for line in text:
    if line.startswith('Review'):
        review.append(line[8:])
    elif line.startswith('Original'):
        original.append(line[24:])
    elif line.startswith('Predicted'):
        predicted.append(line[26:])
        
df['Review'] = review
df['Original'] = original
df['Predicted'] = predicted

df
    
```

```python
from rouge_score import rouge_scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

for index, row in df.iterrows():
    reference = row['Original']
    hypothesis = row['Predicted']
    scores = scorer.score(reference, hypothesis)
    df.at[index, 'rouge1_f1'] = scores['rouge1'].fmeasure
    df.at[index, 'rouge2_f1'] = scores['rouge2'].fmeasure
    df.at[index, 'rougeL_f1'] = scores['rougeL'].fmeasure
```
